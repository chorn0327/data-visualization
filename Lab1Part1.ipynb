```python
import numpy as np
import pandas as pd
import operator
from IPython.display import display, HTML
import math
```

```python
#Question 1
#Read the data into a data frame
print('Data has been read into the dataframe.')
mlb = pd.DataFrame.from_csv('pitching.csv')
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
#Question 2
#Display all pitchers in 2015 under each team sorted by their last names

def getPitchersByYear(year):
    pitchers = mlb.loc[mlb['yearID'] == year]
    pitchers = pitchers.dropna(subset = ['ERA'])
    return pitchers

p2015 = getPitchersByYear(2015)

p2015 = p2015.sort_values(['nameLast'])
display(p2015.loc[:,['yearID','nameFirst', 'nameLast']])

```

```python
#Question 3
# Display all pitchers in 2015 sorted by their ERAs.  This allows you to identify the best and worst pitchers.

p2015byERA = p2015.sort_values(['ERA'])
display(p2015byERA.loc[:,['nameFirst', 'nameLast', 'ERA', 'teamID']])
# print(p2015byERA['nameFirst'] + ' ' + p2015byERA['nameLast'])
```

```python
#Question 4
# Display all pitchers in 2015 under each team sorted by their ERAs
def getVarList(pitchers, var):
    return pitchers[var].unique()



def getTeamPitchers(pitchers):
    teamNames = getVarList(pitchers, 'teamID')
    teamPitchers = {elem: pd.DataFrame for elem in teamNames}
    for key in teamPitchers.keys():
        teamPitchers[key] = pitchers[pitchers.teamID == key]
    return teamPitchers


def displayTeamPitchersInOrderOfVar(teamPitchers, Var):
    for team in teamPitchers.keys():
    #     teamPitchers[team])
        sortedTeam = teamPitchers[team].sort_values([Var])
        display(sortedTeam.loc[:,['teamID','nameFirst', 'nameLast', Var]])
    
teamPitchers = getTeamPitchers(p2015)
displayTeamPitchersInOrderOfVar(teamPitchers, 'ERA')  
```

```python
# Question 5
# Display all pitchers in 2015 under each team sorted by their strike outs (SO)
teamPitchers = getTeamPitchers(p2015)
displayTeamPitchersInOrderOfVar(teamPitchers, 'SO')  
```

```python
# Question 6
# Calculate the average ERA for each team (in 2015), list the teams by the average ERA in a descending order.
def calculateAvgOfColumnForTeams(teamPitchers,Var):
    teamAvg = {}
    for team in teamPitchers.keys():
        teamAvg[team] = sum(teamPitchers[team][Var])
    return teamAvg

teamERAavg = calculateAvgOfColumnForTeams(teamPitchers, 'ERA')
teamERAavg = sorted(teamERAavg.items(), key=operator.itemgetter(1),reverse=True)
print('Average ERA for each team in 2015, listed in descending order')
for team in teamERAavg:
    print (team[0], end=" ")
    print (team[1])
```

```python
# Question 7
# Calculate the total number of home runs allowed (HRs) for each team (in 2015), and list the team by the count in an ascending order
teamHRavg = calculateAvgOfColumnForTeams(teamPitchers, 'HR')
teamHRavg = sorted(teamHRavg.items(), key=operator.itemgetter(1))
print('Average HRs for each team in 2015, listed in ascending order')
for team in teamHRavg:
    print (team[0], end=" ")
    print (team[1])
```

```python
# Question 8
# Calculate each team's average ERA from 2011 to 2015, and list the result as a table, where each column is a year, each row is a team

pitchers1115 = {}
for x in range(2011, 2016):
    pitchers1115[x] = getPitchersByYear(x)


# for p in game.players.passing():
#     d.append({'Player': p, 'Team': p.team, 'Passer Rating':
#         p.passer_rating()})

# pd.DataFrame(d)

teamAvg = {}
f = []
teamlist = getVarList(p2015, "teamID")
for year in pitchers1115.keys():
    teamAvg = calculateAvgOfColumnForTeams(getTeamPitchers(pitchers1115[year]), 'ERA')
    f.append(teamAvg)
    
avgERATab=[['teamID','2011','2012','2013','2014','2015']]

for team in teamlist:
    r = []
    r.append(team)
    for i in range(0,5):
        r.append(f[i].get(team))
    avgERATab.append(r)
print('Average ERA for each team in years 2011 to 2015')
df = pd.DataFrame(avgERATab)
df.columns = df.iloc[0]
df.drop(df.index[0], inplace=True)
pd.options.display.float_format = '{:,.3f}'.format
df.set_index('teamID', inplace=True)
display(df)
```

```python
#Question 9
# Calculate the average ERAs for American League (AL) and National League (NL) from 2011 to 2015, and list the result as a table, where each column is a league and each row is a year

def calculateAvgOfColumnForALNL(teamPitchers,Var):
    teamAvg = {}
    for team in teamPitchers.keys():
        teamAvg[team] = sum(teamPitchers[team][Var])
    return teamAvg

def getALNLPitchers(pitchers):
    alnl = getVarList(pitchers, 'lgID')
    teamPitchers = {elem: pd.DataFrame for elem in alnl}
    for key in teamPitchers.keys():
        teamPitchers[key] = pitchers[pitchers.lgID == key]
    return teamPitchers

pitchers1115 = {}
for x in range(2011, 2016):
    pitchers1115[x] = getPitchersByYear(x)

teamAvg = {}
f = []
leaguelist = getVarList(p2015, "lgID")
for year in pitchers1115.keys():
    teamAvg = calculateAvgOfColumnForALNL(getALNLPitchers(pitchers1115[year]), 'ERA')
    f.append(teamAvg)
    
    


z = []
z.append(['year','AL', 'NL'])
for i in range(0,5):
    z.append([i+2011, f[i].get('AL'), f[i].get('NL') ])



print("Average ERA for each league in years 2011 to 2015")
df = pd.DataFrame(z)
df.columns = df.iloc[0]
df.drop(df.index[0], inplace=True)
pd.options.display.float_format = '{:,.2f}'.format
df.set_index('year', inplace=True)
display(df)
```
